```python
datas1 = datas.copy()
datas1.columns
```

```python
# List of columns to drop
high_correation = ['emp.var.rate', 'euribor3m', 'nr.employed']

# Check which columns from high_corr_cols are present in df1
columns_to_drop = [col for col in high_correation if col in datas1.columns]

# Print columns to be dropped for debugging
print("Columns to be dropped:", columns_to_drop)

# Drop the columns that are present
datas1.drop(columns_to_drop, inplace=True, axis=1)

# Print the remaining columns for verification
print("Remaining columns:", datas1.columns)
```

```python
datas1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
datas_encoded = datas1.apply(lb.fit_transform)
datas_encoded
```

```python
datas_encoded['deposit'].value_counts()
```

```python
x = datas_encoded.drop('deposit',axis=1)  # independent variable
y = datas_encoded['deposit']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split

print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  
    print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
ypred_dt = dt.predict(x_test)
print(ypred_dt)
eval_model(y_test,ypred_dt)
```

```python
from sklearn.tree import plot_tree
```
